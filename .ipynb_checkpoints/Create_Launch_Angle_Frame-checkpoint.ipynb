## Cleaning and pickling statcast data

Import statcast data from csv, clean data and pickle for later use.

```python
from __future__ import print_function, division
from types import *

import numpy as np
import pandas as pd
import pickle
import re

```

Read batted ball data for years 2015, 2016, and 2017

```python
def read_ball_data(file):
    
    return pd.read_csv(file,low_memory=False)

def create_frame(file_names):

    frames = []
    for file in file_names:
        print(file)
        frames.append(read_ball_data(file))
        
    return pd.concat(frames)    
        

file_names = ['savant_data.csv','savant_data_1.csv','savant_data_2.csv',
              'savant_data_3.csv', 'savant_data_4.csv',
              'savant_data_5.csv']            

savant_data = create_frame(file_names)
    
```

```python
df = savant_data.copy()
```

```python
df.shape
```

Read year of event from date string

```python
def read_year(string):
    if re.match(r'2017-[0-9]+-[0-9]+',string):
        print(string)
        return 2017
    if re.match(r'2016-[0-9]+-[0-9]+',string):
        return 2016
    if re.match(r'2015-[0-9]+-[0-9]+',string):
        return 2015
    else:
        return np.NaN

df['Year'] = df['game_date'].map(lambda x: read_year(x))
```

Create a dataframe using only desired data

```python
df[df.Year == np.NaN]
```

```python
df1 = df[['player_name','Year','events','launch_speed','launch_angle']]
```

```python
df1.rename(columns = {'player_name':'Name'},inplace = True)
```

```python
df1.shape
```

```python
df1[['Year','launch_speed','launch_angle']] = df1[['Year','launch_speed','launch_angle']].apply(pd.to_numeric,errors='coerce')
```

```python
df2 = df1.loc[df1.launch_angle != np.floor(df1.launch_angle)].copy()
```

```python
df2.shape
```

```python
df2[df2['Year'] == 2017].describe()
```

```python
def count_bases(event):
    
    if event =='single':
        return 1             
    if event =='double':
        return 2             
    if event =='triple':
        return 3
    if event =='home_run':
        return 4
    else:
        return 0
    
df2['num_bases'] = df2['events'].map(lambda x: count_bases(x))
```

```python
def count_hit(event):
    
    if event in ['single','double','triple','home_run']:
        return 1             
    else:
        return 0
    
df2['hit'] = df2['events'].apply(count_hit)
```

```python
def count_HR(event):
    
    if event == 'home_run':
        return 1             
    else:
        return 0

    
df2['HR'] = df2['events'].apply(count_HR)
```

```python
df2[(df2['Name'] == 'Daniel Murphy') & (df2['Year'] == 2016)].describe()
```

```python
df2.to_pickle('event.pkl')
```

```python
ball_data = df2.groupby(['Name','Year']).mean().reset_index()
```

```python
def year_change(df):
    prev_df = df.copy()
    prev_df['Year'] = prev_df['Year'].apply(lambda x: x+1)    #str(int(prev_df.Year)+1)
    prev_df = prev_df.rename(columns = {'launch_speed':'prev_launch_speed',
                        'launch_angle':'prev_launch_angle',
                        'num_bases':'prev_num_bases',
                        'hit':'prev_hit',
                        'HR':'prev_HR'
                             })
    
    new_df = pd.merge(df,prev_df,how = 'left',on=['Name','Year'])
    
    return new_df

ball_data = year_change(ball_data)
#event_data['Year'] = event_data['Year'].apply(pd.to_numeric)
```

```python
#add deltas

def deltas(df):
    
    df['change_angle'] = df['launch_angle'] - df['prev_launch_angle']
    df['change_speed'] = df['launch_speed'] - df['prev_launch_speed']
    df['change_SLG'] = df['num_bases'] - df['prev_num_bases']
    df['change_hit'] = df['hit'] - df['prev_hit']
    df['change_HR'] = df['HR'] - df['prev_HR']
    
    return

deltas(ball_data)
```
